```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cross_validation import cross_val_score
from sklearn.metrics import auc


n_samples = 10000

X = pd.read_csv("./data/algomostchem_train.txt", header=None, index_col=None)
Y = pd.read_csv("./data/algomostchem_trainY.txt", header=None, index_col=None, names=['target'])

indexes = np.random.random_integers(0, len(X), n_samples)

X = X.iloc[indexes, :]
Y = Y.iloc[indexes, :]

np.random.seed(60115)
```

```python
R = X.corr()
plt.pcolor(R)
plt.colorbar()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# вычтем среднее из X чтобы отцентровать выборку
scaler = StandardScaler(with_std=False).fit(X)
X_scaled = scaler.transform(X)
# запустим PCA с 3мя основным компонентами, чтобы посмотреть что из этого получится
X_reduced = PCA(n_components=3).fit_transform(X_scaled)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Нарисуем то что получилось в пространстве 3х главных компонент
plt.ion()



fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-100, azim=110)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=Y.as_matrix(),
           cmap=plt.cm.Paired)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# т.к. наверху явно видно 3 каких-то сгустка. Попробуем разделить Х на 3 кластера и нарисовать каждый кластер отдельно
# у меня нет логического обхяснения зачем это надо, но звучит прикольно, и может что-то полезное удастся разглядеть

kmeans = KMeans(n_clusters=3, n_jobs=-1)

kmeans.fit(X_reduced)

# меняя 0 на 1 или 2 можно рисовать разные сгустки
X2 = X_reduced[kmeans.labels_ == 0]
Y2 = Y.as_matrix().reshape(n_samples)[kmeans.labels_ == 0]

fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=-150)
ax.scatter(X2[:, 0], X2[:, 1], X2[:, 2], c=Y2,
           cmap=plt.cm.Paired)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```

```python
import matplotlib.pyplot as plt

# обучим gbm с заведомо большим числом деревьев, чтобы потом оценить оптимальное число деревьев (см. ячейку ниже)
# кстати делать это будем в пространстве главных компонент, объясняющих 99.9 % дисперсии
pca = PCA(n_components=0.999)
X_reduced = pca.fit_transform(X_scaled)

n_trees = 200

gbc = GradientBoostingClassifier(n_estimators=n_trees, subsample=.8)
gbc.fit(X_reduced, y=Y.as_matrix().reshape(n_samples))
```

```python
oob_color = list(map(lambda x: x / 256.0, (190, 174, 212)))
test_color = list(map(lambda x: x / 256.0, (127, 201, 127)))

# построим графики ошибки на обучении и Out-of-Bag 
# по мотивам: http://scikit-learn.org/stable/auto_examples/ensemble/plot_gradient_boosting_oob.html
iterations = np.arange(n_trees) + 1

plt.plot(iterations, gbc.train_score_, label='Train Error', color=test_color)
plt.plot(iterations, -np.cumsum(gbc.oob_improvement_), label='Out-of-Bag Error', color=oob_color)
```
