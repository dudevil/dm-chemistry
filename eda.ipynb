## GBM

```python
import matplotlib.pyplot as plt

# обучим gbm с заведомо большим числом деревьев, чтобы потом оценить оптимальное число деревьев (см. ячейку ниже)
# кстати делать это будем в пространстве главных компонент, объясняющих 99.9 % дисперсии
pca = PCA(n_components=0.999)
X_reduced = pca.fit_transform(X_scaled)

n_trees = 200
oob_color = list(map(lambda x: x / 256.0, (190, 174, 212)))
test_color = list(map(lambda x: x / 256.0, (127, 201, 127)))

gbc = GradientBoostingClassifier(n_estimators=n_trees, subsample=.8)
gbc.fit(X_reduced, y=Y.as_matrix().reshape(n_samples))

# построим графики ошибки на обучении и Out-of-Bag 
# по мотивам: http://scikit-learn.org/stable/auto_examples/ensemble/plot_gradient_boosting_oob.html
iterations = np.arange(n_trees) + 1
oob_scores = -np.cumsum(gbc.oob_improvement_)
best_trees = np.argmin(oob_scores)

plt.plot(iterations, gbc.train_score_, label='Train Error', color=test_color)
plt.plot(iterations, oob_scores, label='Out-of-Bag Error', color=oob_color)

print("Best ot-of-bag error: %f with %d trees" % (oob_scores[best_trees], best_trees))
```

## Separate models for clusters

Идея этой секции состоит в том, чтобы попробовать разделить выборку на 3 кластера (см. разведывательный анализ выше) и попробовать обучить модельки на каждом из классов отдельно

```python
from sklearn.cluster import KMeans
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.cross_validation import cross_val_score

X_scaled = scale(X)

n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, n_jobs=-1, random_state=np.random.seed(60115))
kmeans.fit(X_scaled)

cv_clusters = np.zeros(n_clusters, dtype=np.float)

for i in xrange(n_clusters):
    lreg = LogisticRegression(class_weight='auto')
    scores = cross_val_score(lreg, 
                             X_scaled[kmeans.labels_ == i],
                             Y.iloc[kmeans.labels_ == i].values.flatten(), 
                             scoring='roc_auc', 
                             cv=5, n_jobs=3)
    cv_clusters[i] = scores.mean() * sum(kmeans.labels_ == i)
    print("CV for cluster %d: %f with std: %f" % (i + 1, scores.mean(), scores.std()))
    
print("Weighted CV average: %f" % (np.sum(cv_clusters) / len(X_scaled)))
```
