```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cross_validation import cross_val_score
from sklearn.metrics import auc


n_samples = 10000

X = pd.read_csv("./data/algomostchem_train.txt", header=None, index_col=None)
Y = pd.read_csv("./data/algomostchem_trainY.txt", header=None, index_col=None, names=['target'])

indexes = np.random.random_integers(0, len(X), n_samples)

X = X.iloc[indexes, :]
Y = Y.iloc[indexes, :]

np.random.seed(60115)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# вычтем среднее из X чтобы отцентровать выборку
scaler = StandardScaler(with_std=False).fit(X)
X_scaled = scaler.transform(X)
# запустим PCA с 3мя основным компонентами, чтобы посмотреть что из этого получится
X_reduced = PCA(n_components=3).fit_transform(X_scaled)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Нарисуем то что получилось в пространстве 3х главных компонент
plt.ion()



fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-100, azim=110)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=Y.as_matrix(),
           cmap=plt.cm.Paired)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# т.к. наверху явно видно 3 каких-то сгустка. Попробуем разделить Х на 3 кластера и нарисовать каждый кластер отдельно
# у меня нет логического обхяснения зачем это надо, но звучит прикольно, и может что-то полезное удастся разглядеть

kmeans = KMeans(n_clusters=3, n_jobs=-1)

kmeans.fit(X_reduced)

# меняя 0 на 1 или 2 можно рисовать разные сгустки
X2 = X_reduced[kmeans.labels_ == 0]
Y2 = Y.as_matrix().reshape(n_samples)[kmeans.labels_ == 0]

fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=-150)
ax.scatter(X2[:, 0], X2[:, 1], X2[:, 2], c=Y2,
           cmap=plt.cm.Paired)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# обучим градиентный бустинг и посмотрим распределение фич по "важности" с точки срения gbm

gbc = GradientBoostingClassifier(n_estimators=50)
gbc.fit(X_scaled, Y.values.flatten())
plt.hist(gbc.feature_importances_, bins=30)
```

## SVM

```python
from sklearn import svm


# попытки поиграть с SVM, лучше даже не запускать: время только тратить

#Cs = np.logspace(-5,5,10)
C = 1e4
gammas = 10. ** np.arange(-1,5)
# array([  1.00000000e-01,   1.00000000e+00,   1.00000000e+01,
#         1.00000000e+02,   1.00000000e+03,   1.00000000e+04])

for gamma in gammas:
    svm_class = svm.SVC(C=C, gamma=gamma)
    scores = cross_val_score(svm_class, X_scaled, y=Y.values.flatten(), scoring='roc_auc', cv=3)
    print("gamma=%f,\t score: %f" % (gamma, scores.mean()))
```

## ExtraTreesClassifier

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.cross_validation import cross_val_score
from sklearn.metrics import auc
from sklearn.decomposition import PCA

# Это что-то вроде RandomForest'a

#pca = PCA(n_components=0.999)
#X_reduced = pca.fit_transform(X_scaled)

n_trials = 15

n_trees = np.logspace(1, 4, n_trials, dtype=np.int)
cv_scores = np.zeros(n_trials, dtype=np.float)
#oob_scores = np.zeros(n_trials, dtype=np.float)

for i, n_estimators in enumerate(n_trees):
    etc = ExtraTreesClassifier(n_estimators=n_estimators, n_jobs=-1, bootstrap=True)
    cv_scores[i] = cross_val_score(etc, X_scaled, y=Y.values.flat, scoring='roc_auc', cv=5).mean()

plt.plot(n_trees, cv_scores)
print('Best CV-score: %f with %d trees.' % (np.max(cv_scores), n_trees[np.argmax(cv_scores)]))
```

## GBM

```python
import matplotlib.pyplot as plt

# обучим gbm с заведомо большим числом деревьев, чтобы потом оценить оптимальное число деревьев (см. ячейку ниже)
# кстати делать это будем в пространстве главных компонент, объясняющих 99.9 % дисперсии
pca = PCA(n_components=0.999)
X_reduced = pca.fit_transform(X_scaled)

n_trees = 200
oob_color = list(map(lambda x: x / 256.0, (190, 174, 212)))
test_color = list(map(lambda x: x / 256.0, (127, 201, 127)))

gbc = GradientBoostingClassifier(n_estimators=n_trees, subsample=.8)
gbc.fit(X_reduced, y=Y.as_matrix().reshape(n_samples))

# построим графики ошибки на обучении и Out-of-Bag 
# по мотивам: http://scikit-learn.org/stable/auto_examples/ensemble/plot_gradient_boosting_oob.html
iterations = np.arange(n_trees) + 1
oob_scores = -np.cumsum(gbc.oob_improvement_)
best_trees = np.argmin(oob_scores)

plt.plot(iterations, gbc.train_score_, label='Train Error', color=test_color)
plt.plot(iterations, oob_scores, label='Out-of-Bag Error', color=oob_color)

print("Best ot-of-bag error: %f with %d trees" % (oob_scores[best_trees], best_trees))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.cross_validation import cross_val_score
from sklearn.metrics import auc
from sklearn.decomposition import PCA

n_trials = 30
pca = PCA(n_components=0.999)
X_reduced = pca.fit_transform(X_scaled)

n_trees = np.linspace(30, 90, n_trials, dtype=np.int)
cv_scores = np.zeros((n_trials, 2), dtype=np.float)

for i, n_estimators in enumerate(n_trees):
    gbc = GradientBoostingClassifier(n_estimators=n_estimators)
    scores = cross_val_score(gbc, X_reduced, y=Y.values.flatten(), scoring='roc_auc', cv=5)
    cv_scores[i, 0] = scores.mean()
    cv_scores[i, 1] = scores.std()
    
#plt.plot(neighbors, cv_scores[:, 0])
plt.errorbar(n_trees, cv_scores[:, 0], yerr=cv_scores[:, 1])
print('Best CV-score: %f with %d trees.' % (np.max(cv_scores[:, 0]), n_trees[np.argmax(cv_scores[:, 0])]))
```

## K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import scale

n_trials = 30
# дисперсию тоже стандартизуем
X_std = scale(X_scaled)

neighbors = np.linspace(3, 50, n_trials, dtype=np.int)
cv_scores = np.zeros((n_trials, 2), dtype=np.float)
    
for i, n_neighbors in enumerate(neighbors):
    knnc = KNeighborsClassifier(n_neighbors=n_neighbors)
    scores = cross_val_score(knnc, X_scaled, y=Y.values.flatten(), scoring='roc_auc', cv=5)
    cv_scores[i, 0] = scores.mean()
    cv_scores[i, 1] = scores.std()
    
#plt.plot(neighbors, cv_scores[:, 0])
plt.errorbar(neighbors, cv_scores[:, 0], yerr=cv_scores[:, 1])
print('Best CV-score: %f with %d neighbors.' % (np.max(cv_scores[:, 0]), neighbors[np.argmax(cv_scores[:, 0])]))
```
